# I-JEPA on OpenBHB with NIDL

This notebook will show you how to use [I-JEPA model](https://arxiv.org/pdf/2301.08243) on OpenBHB. It is a self-supervised model that relies on mask modeling in the latent space of a vision transformer for training. It does **not** require strong data augmentation as in contrastive learning however it needs a Transformer-like architecture for encoding the images.

```python
from nidl.datasets.openbhb import OpenBHB
from nidl.models import JEPA
from nidl.volume.transforms import CropOrPad
from nidl.data.collate import MaskCollateFunction
from torch.utils.data import DataLoader
```

## Load quasi-raw data and fit I-JEPA model

```python
# Load the data
dataset_train = OpenBHB("/neurospin/signatures/bd261576/openBHB", target="age", modality="quasiraw", split="train", 
                        transforms=CropOrPad((1, 128, 128, 128)))
dataset_test = OpenBHB("/neurospin/signatures/bd261576/openBHB", target="age", modality="quasiraw", split="val",
                       transforms=CropOrPad((1, 128, 128, 128)))
train_dataloader = DataLoader(dataset_train, batch_size=4, num_workers=15, shuffle=True, 
                              collate_fn=MaskCollateFunction())
val_dataloader = DataLoader(dataset_test, batch_size=4, num_workers=10, shuffle=False,
                            collate_fn=MaskCollateFunction())
```

```python
jepa = JEPA(
    encoder="vit_tiny_3d",
    encoder_kwargs={"pool": None},
    predictor_kwargs={"predictor_embed_dim": 192, "num_heads": 3, "depth": 6},
    lr_scheduler=None,
    num_sanity_val_steps=0
)
```

```python
jepa.fit(train_dataloader, val_dataloader)
```
